```python
X = wine.drop(['quality'], axis =1)
y = wine['quality']

print('Shape of X', X.shape)
print('Shape of y', y.shape)
```

```python
# scalling? menyeleksi feature yg bertipe numerik, tapi ini numerik semua. jd perlu g? g

from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
standard_scaler.fit_transform(wine)


```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
def models(X_train, y_train):

  #Decision Tree
  from sklearn.tree import DecisionTreeClassifier
  tree = DecisionTreeClassifier()
  tree.fit(X_train, y_train)
  y_pred = tree.predict(X_test)

  print('f1', classification_report(y_test, y_pred))
  print('matrix',confusion_matrix(y_test, y_pred))
  print('acc',accuracy_score(y_test, y_pred))


  #Naive Bayes 

  from sklearn.naive_bayes import GaussianNB
  nb = GaussianNB()
  nb.fit(X_train, y_train)
  y_pred = nb.predict(X_test)

  print('f1', classification_report(y_test, y_pred))
  print('matrix',confusion_matrix(y_test, y_pred))
  print('acc',accuracy_score(y_test, y_pred))

  #KNeighbors 

  from sklearn.neighbors import KNeighborsClassifier
  neighbors = KNeighborsClassifier()
  neighbors.fit(X_train, y_train)
  y_pred = neighbors.predict(X_test)

  print('f1', classification_report(y_test, y_pred))
  print('matrix', confusion_matrix(y_test, y_pred))
  print('acc', accuracy_score(y_test, y_pred))

  print('Decision Tree acc: ', tree.score(X_train, y_train))
  print('Naive Bayes acc: ', nb.score(X_train, y_train))
  print('KNeighbors acc: ', neighbors.score(X_train, y_train))

  return tree, nb, neighbors


```

```python
model =  models(X_train, y_train)
```
